Составьте программу для нахождения всех натуральных решений
уравнения x2+y2+z2=k2
, где x,y,z∈[1,30], а k вводится с клавиатуры

```python
# Задание task_02_28.
#
# Выполнил: Щукин М.А.
# Группа: ЦИБ-251



k = int(input('k = '))

for x in range(1, 31):
  for y in range(1, 31):
    for z in range(1, 31):
      if x ** 2 + y ** 2 + z ** 2 == k ** 2:
        print(f"x = {x}, y = {y}, z = {z}")
```
